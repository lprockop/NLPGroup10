```python
API_KEY='aLSPI0NgWKKQ7If0au0DnU2tMpF9opR0' #insert your API_KEY 

books = []
for i in range(10):
  url='https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key='+API_KEY
  response = req.get(url).json()
  print(response)

  docs = response['results']['books']
  for doc in docs:
      filteredDoc = {}
      filteredDoc['rank'] = doc['rank']
      filteredDoc['rank_last_week'] = doc['rank_last_week']
      filteredDoc['title']=doc['title']
      filteredDoc['author']=doc['author']
      filteredDoc['desc']=doc['description']
      filteredDoc['publisher']=doc['publisher']

      books.append(filteredDoc)

books[:10]

# can then export to csv using pandas
```

```python
rank_list = []
for a, i in enumerate(books):
  rank_list.append(i['rank'])

rank_last_week_list = []
for a, i in enumerate(books):
  rank_last_week_list.append(i['rank_last_week'])

import numpy as np 
xax = np.arange(0, 150, 1)

diff = []
for i in xax:
  diff_tmp = rank_list[i] - rank_last_week_list[i]
  diff.append(diff_tmp)
```

```python
import matplotlib.pyplot as plt

plt.scatter(rank_list, rank_last_week_list, label='last week rank', color='lightgreen', alpha=0.5)
plt.scatter(rank_list, diff, label='movement since last week', color='yellow', alpha=0.5)
plt.xlabel('current rank')
plt.legend()
plt.show()
```
